# House Price Analysis

## Bivariate and Multivariate Analysis

### Objective
To analyze the factors affecting house prices and identify relationships between area, BHK count, location, and property price.




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv("train.csv")
```

```python
sns.scatterplot(x='SQUARE_FT',y='TARGET(PRICE_IN_LACS)',data = df)
plt.show()               
```

```python
df['SQUARE_FT'].describe()
```

```python
df.nlargest(10, 'SQUARE_FT')[['SQUARE_FT', 'TARGET(PRICE_IN_LACS)']]

sns.boxplot(
    x='BHK_NO.',
    y='TARGET(PRICE_IN_LACS)',
    data=df
)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='SQUARE_FT',
    y='TARGET(PRICE_IN_LACS)',
    hue='BHK_NO.',
    data=df)

plt.title("Area vs Price by BHK")
plt.show()
```
