## Introduction

## About the Dataset

The Mall Customer Dataset contains information about customers visiting a shopping mall.

### Features Used

- Age
- Annual Income (k$)
- Spending Score (1-100)

---

## What is Standardization?

Standardization is a feature scaling technique that transforms data so that it has:

- Mean ≈ 0
- Standard Deviation ≈ 1

It is useful when features have different scales.

### Formula

Z = (X - μ) / σ

Where:

- X = Original value
- μ = Mean
- σ = Standard Deviation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
num_cols = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']

df_num = df[num_cols]
```

```python
df_num.describe()
```

```python
df_num.hist(figsize=(12,5))
plt.suptitle("Before Standardization")
plt.show()
```

```python
scaler = StandardScaler()

scaled_data = scaler.fit_transform(df_num)

df_scaled = pd.DataFrame(scaled_data, columns=num_cols)

df_scaled.head()
```

```python
df_scaled.describe()
```

```python
df_scaled.mean()
```

```python
df_scaled.std()
```

```python
df_scaled.hist(figsize=(12,5))
plt.suptitle("After Standardization")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df_num)
plt.title("Before Standardization")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df_scaled)
plt.title("After Standardization")
plt.show()
```

```python
comparison = pd.DataFrame({
    'Original Mean': df_num.mean(),
    'Standardized Mean': df_scaled.mean(),
    'Original Std': df_num.std(),
    'Standardized Std': df_scaled.std()
})

comparison
```

```python
plt.figure(figsize=(10,6))

for col in df_num.columns:
    sns.kdeplot(df_num[col], label=col, fill=True, alpha=0.3)

plt.title("KDE Plot Before Standardization")
plt.xlabel("Values")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))

for col in df_scaled.columns:
    sns.kdeplot(df_scaled[col], label=col, fill=True, alpha=0.3)

plt.title("KDE Plot After Standardization")
plt.xlabel("Standardized Values")
plt.ylabel("Density")
plt.legend()
plt.show()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 10))

for i, col in enumerate(num_cols):
    sns.kdeplot(df_num[col], ax=axes[i,0], fill=True)
    axes[i,0].set_title(f"{col} - Before")

    sns.kdeplot(df_scaled[col], ax=axes[i,1], fill=True)
    axes[i,1].set_title(f"{col} - After")

plt.tight_layout()
plt.show()
```

## What is Min-Max Normalization?

Min-Max Normalization is a feature scaling technique that transforms data into a fixed range, typically between 0 and 1.

After normalization:

- Minimum value becomes 0
- Maximum value becomes 1

This technique preserves the original distribution of the data while changing its scale.

### Formula

X_normalized = (X - X_min) / (X_max - X_min)

Where:

- X = Original value
- X_min = Minimum value of the feature
- X_max = Maximum value of the feature

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("winequality-red.csv", sep=';')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
df.describe()
df.isnull().sum()
num_cols = df.columns
df_num = df[num_cols]
```

```python
scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(df_num)

df_normalized = pd.DataFrame(normalized_data, columns=num_cols)

df_normalized.head()
```

```python
df_normalized.min()
```

```python
df_normalized.max()
```

```python
plt.figure(figsize=(12,6))

for col in df_normalized.columns:
    sns.kdeplot(df_normalized[col], label=col)

plt.title("KDE Plot After Normalization")
plt.xlabel("Normalized Values")
plt.ylabel("Density")
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(12,10))

for col in df_num.columns:
    sns.kdeplot(df_num[col], ax=axes[0])

axes[0].set_title("Before Normalization")

for col in df_normalized.columns:
    sns.kdeplot(df_normalized[col], ax=axes[1])

axes[1].set_title("After Normalization")

plt.tight_layout()
plt.show()
```

```python
comparison = pd.DataFrame({
    'Original Min': df_num.min(),
    'Normalized Min': df_normalized.min(),
    'Original Max': df_num.max(),
    'Normalized Max': df_normalized.max()
})

comparison
```
