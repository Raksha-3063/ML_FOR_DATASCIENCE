# Netflix Dataset Analysis using Pandas Profiling

## Objective
The objective of this project is to perform Exploratory Data Analysis (EDA) on the Netflix dataset using Pandas and YData Profiling. The analysis helps in understanding the dataset structure, identifying missing values, handling data quality issues, and generating an automated profiling report.

---

## Dataset Overview
The Netflix dataset contains information about movies and TV shows available on Netflix, including:

- Show ID
- Type (Movie/TV Show)
- Title
- Director
- Cast
- Country
- Date Added
- Release Year
- Rating
- Duration
- Genre
- Description

---

## Steps Performed

### 1. Imported Required Libraries
- Pandas for data manipulation and analysis.
- YData Profiling for automated exploratory data analysis.

### 2. Loaded the Dataset
The dataset was loaded into a Pandas DataFrame and the first few records were examined to understand its structure.

### 3. Explored the Dataset
The following checks were performed:
- Dataset dimensions (rows and columns)
- Column names
- Data types
- Missing values

### 4. Missing Value Analysis
The following missing values were identified:

| Column | Missing Values |
|----------|----------|
| director | 2634 |
| cast | 825 |
| country | 831 |
| date_added | 10 |
| rating | 4 |
| duration | 3 |

### 5. Handling Missing Values

#### Director, Cast, and Country
Since these columns contain categorical information, missing values were replaced with **"Unknown"**.

#### Rating
Missing values were replaced using the **Mode** (most frequently occurring value).

#### Duration
Missing values were replaced using the **Mode** of the column.

#### Date Added
Missing values were replaced using the **Mode** of the column.

### 6. Verification
After data cleaning, the dataset was checked again and all missing values were successfully handled.

### 7. Pandas Profiling Report Generation
An automated profiling report was generated to analyze:

- Dataset overview
- Missing values
- Variable distributions
- Correlations
- Duplicate records
- Data types
- Statistical summaries

### 8. Report Export
The profiling report was exported as an HTML file for future reference and sharing.

---

## Key Learnings

- Understanding dataset structure using Pandas.
- Identifying and handling missing values.
- Using Mode for categorical data imputation.
- Generating automated EDA reports using YData Profiling.
- Interpreting dataset statistics and quality metrics.

---

## Conclusion

The Netflix dataset was successfully analyzed and cleaned using Pandas. Missing values were handled appropriately, and a comprehensive profiling report was generated using YData Profiling. This process provided valuable insights into the dataset and prepared it for further visualization and machine learning tasks.

```python
import pandas as pd
from ydata_profiling import ProfileReport
df = pd.read_csv("netflix_titles.csv")
```

```python
from ydata_profiling import ProfileReport

profile = ProfileReport(df, explorative=True)
profile.to_notebook_iframe()
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
```

```python
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
df['duration'] = df['duration'].fillna(df['duration'].mode()[0])
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
```
